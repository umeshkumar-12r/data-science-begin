```python
import pandas as pd
```

```python
data1 = {
    "name":["umesh","prince","vikas"],
    "english":[29,28,29],
    "hindi":[29,29,30]
}
data = pd.DataFrame(data1)
```

```python
print(data)
```

```python
data["average"] = (data["english"] + data["hindi"])/2
```

```python
data["result"] = data["average"].apply(lambda x: "pass" if x > 20 else "fail")
```

```python
print(data)
```

```python
group = data.groupby("result")
```

```python
print(group.size())
```

```python
print(group["english"].mean())
```

```python
group["hindi"].agg(["mean","count","max","min"])
```
