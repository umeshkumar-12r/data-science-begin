```python
import numpy as np
marks = np.array([12,17,19,20,14])
print(marks)
```

```python
mean_of_marks = np.mean(marks)
print(mean_of_marks)
```

```python
max_mark = max(marks)
print(max_mark)
```

```python
updated_marks = marks +5
print(updated_marks)
```
